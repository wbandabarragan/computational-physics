# Partícula cuántica en un potencial infinito 2D

Consideramos una partícula de masa $m$ que se mueve en un espacio rodeado por barreras infinitas impenetrables.

El potencial, $V(x)$, está dado por la siguiente función a trozos:

$$
V(x,y)= 
\begin{cases}
0, & 0 \leq x \leq L_x, 0 \leq y \leq L_y \\
\infty, & \text{en otros $x,y$}
\end{cases}
$$

Dado que $V(x)$ es independiente del tiempo, las soluciones de la ecuación de Schrödinger se pueden separar en funciones de onda espaciales y temporales:

$$
\Psi(x,y,t) = A\sin\left(\frac{n_x\pi x}{L_x}\right)\sin\left(\frac{n_y\pi y}{L_y}\right)\,e^{-i\frac{E_n}{\hbar}t}
$$

El último término temporal es el denominado "factor de oscilación".

Vamos a representar gráficamente las soluciones estacionarias utilizando herramientas de Python.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as ct
```

```python
# Estados estacionarios

def psi_function_2D(l_x, l_y, n_x, n_y, x_2D, y_2D, t):
    """
    Función utilizada para calcular los estados estacionarios de una partícula cuántica 2D en un potencial infinito.
    """
    # Cálculo de Energías
    E_n = (n_x**2*np.pi**2*ct.hbar**2)/(2*m*l_x**2) + (n_y**2*np.pi**2*ct.hbar**2)/(2*m*l_y**2)

    # Cómputo de las Funciones de Onda
    f_xy = np.sin(n_x*np.pi*x_2D/l_x)*np.sin(n_y*np.pi*y_2D/l_y)*np.cos(E_n*t/ct.hbar)
    return f_xy
```

```python
# Condiciones iniciales
l_x = l_y = 2.e-10
m = ct.m_e

# Ejes
x_axis = np.arange(-3.e-10, 5.e-10, 5.e-13)
y_axis = np.arange(-3.e-10, 5.e-10, 5.e-13)

# Mallado 2D
x_2D, y_2D = np.meshgrid(x_axis, y_axis)

#print(x_2D.shape, y_2D.shape)
```

```python
# Prueba de concepto a un tiempo determinado t=0: 4 estados estacionarios

y1 = psi_function_2D(l_x, l_y, 1, 1, x_2D, y_2D, 0) # n=1

y2 = psi_function_2D(l_x, l_y, 2, 2, x_2D, y_2D, 0) # n=2

y3 = psi_function_2D(l_x, l_y, 3, 3, x_2D, y_2D, 0) # n=3

y4 = psi_function_2D(l_x, l_y, 4, 4, x_2D, y_2D, 0) # n=4
```

```python
# Inicia la figura
fig, ax = plt.subplots(2,2, figsize=(10,10))

ax[0][0].pcolor(x_2D, y_2D, y1,  label = r'$\psi_1$', clim=(-1., +1.))
ax[0][1].pcolor(x_2D, y_2D, y2,  label = r'$\psi_2$', clim=(-1., +1.))
ax[1][0].pcolor(x_2D, y_2D, y3,  label = r'$\psi_3$', clim=(-1., +1.))
ax[1][1].pcolor(x_2D, y_2D, y4,  label = r'$\psi_4$', clim=(-1., +1.))

#Z1 = plt.pcolor(x_2D, y_2D, y1)

#ax[0][0].colorbar(Z1)

ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

ax[0][0].set_xlim(0,2.e-10)
ax[0][1].set_xlim(0,2.e-10)
ax[1][0].set_xlim(0,2.e-10)
ax[1][1].set_xlim(0,2.e-10)

ax[0][0].set_ylim(0,2.e-10)
ax[0][1].set_ylim(0,2.e-10)
ax[1][0].set_ylim(0,2.e-10)
ax[1][1].set_ylim(0,2.e-10)

plt.show()
```

```python
for i in np.arange(0,5*4.3992935759667446e-16,4.3992935759667446e-18):
    
    y1 = psi_function_2D(l_x, l_y, 1, 1, x_2D, y_2D, i) # n=1

    y2 = psi_function_2D(l_x, l_y, 2, 2, x_2D, y_2D, i) # n=2

    y3 = psi_function_2D(l_x, l_y, 3, 3, x_2D, y_2D, i) # n=3

    y4 = psi_function_2D(l_x, l_y, 4, 4, x_2D, y_2D, i) # n=4
    
    fig, ax = plt.subplots(2,2, figsize=(10,10))

    ax[0][0].pcolor(x_2D, y_2D, y1,  label = r'$\psi_1$', clim=(-1., +1.))
    ax[0][1].pcolor(x_2D, y_2D, y2,  label = r'$\psi_2$', clim=(-1., +1.))
    ax[1][0].pcolor(x_2D, y_2D, y3,  label = r'$\psi_3$', clim=(-1., +1.))
    ax[1][1].pcolor(x_2D, y_2D, y4,  label = r'$\psi_4$', clim=(-1., +1.))

    #Z1 = plt.pcolor(x_2D, y_2D, y1)

    #ax[0][0].colorbar(Z1)

    ax[0][0].legend()
    ax[0][1].legend()
    ax[1][0].legend()
    ax[1][1].legend()

    ax[0][0].set_xlim(0,2.e-10)
    ax[0][1].set_xlim(0,2.e-10)
    ax[1][0].set_xlim(0,2.e-10)
    ax[1][1].set_xlim(0,2.e-10)

    ax[0][0].set_ylim(0,2.e-10)
    ax[0][1].set_ylim(0,2.e-10)
    ax[1][0].set_ylim(0,2.e-10)
    ax[1][1].set_ylim(0,2.e-10)
    
    j = i*1.e17

    #plt.show()
    plt.savefig("wave2D_functions{:07.2f}.png".format(j))
    plt.close()
```
